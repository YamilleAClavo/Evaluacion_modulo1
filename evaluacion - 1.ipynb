```python
class TiendaOnline:
    def __init__(self, ventas_totales = float(0)): #Se definen los atributos de la clase y se especifica el valor de ventas totales que es tipo float ya que sabemos que en principio es 0
        self.inventario = [] #Lista vacia que almacena los productos
        self.clientes = {} #Diccionario vacio que almacena los clientes
        self.ventas_totales = ventas_totales

    #definimos los métodos asociados a la clase TiendaOnline    

    def agregar_producto(self,nombre, precio, cantidad):
        diccionario_producto = {"nombre" : nombre, "precio": precio, "cantidad" : cantidad} # creamos un diccionario con nombre, precio y cantidad
        for diccionario_elemento in self.inventario: #iteramos sobre cada diccionario de la lista
            if diccionario_elemento["nombre"] == nombre: #comprobamos si el producto existe
               diccionario_elemento['cantidad'] += cantidad #sumamos la cantidad
               print(f"producto {nombre} ya existe, actualización de cantidad realizada con éxito: {diccionario_elemento['cantidad']} unidades")
               return self.inventario
        
        self.inventario.append(diccionario_producto) #si el producto no existe lo añadimos 
        print(f"El producto {nombre} ha sido agregado correctamente")
        return self.inventario
    
    def ver_inventario(self):
        for diccionario_elemento in self.inventario: # iteramos sobre cada diccionario de la lista
        
            print(f"Nombre: {diccionario_elemento['nombre']}, Precio:{diccionario_elemento ['precio']}, Cantidad:{diccionario_elemento['cantidad']}") 

    
    def buscar_producto(self, nombre):
        for diccionario_elemento in self.inventario: #recorremos cada diccionario de la lista
            if diccionario_elemento["nombre"] == nombre: #buscamos el producto en cada diccionario y si lo encontramos printeamos la información
                print(f"Nombre: {diccionario_elemento['nombre']}, Precio:{diccionario_elemento ['precio']}, Cantidad:{diccionario_elemento['cantidad']}")
                
                return self.inventario 
        print(f"El producto {nombre} no existe en el inventario actual")

    def actualizar_stock(self,nombre, cantidad):
        for diccionario_elemento in self.inventario: #recorremos los diccionarios de la lista
            if diccionario_elemento ["nombre"] == nombre: #buscamos coincidencia del producto en los diccionarios de la lista
                diccionario_elemento["cantidad"] += cantidad #sumamos
                return self.inventario
            
        print(f"El producto {nombre} no existe en el inventario actual")

    def eliminar_producto(self, nombre):
        for diccionario_elemento in self.inventario:
            if diccionario_elemento["nombre"] == nombre:
                self.inventario.remove(diccionario_elemento) #si encontramos el elemento lo eliminamos
                print(f"el producto {nombre} ha sido eliminado del inventario")
                return self.inventario
        print(f"El producto {nombre} no existe en el inventario actual")

    def calcular_valor_inventario(self):
        valor_total = 0 #variable que iremos sumando en cada iteración
        for diccionario_elemento in self.inventario:
            valor_producto = diccionario_elemento["cantidad"] * diccionario_elemento["precio"] # variable para saber el valor de cada producto: unidades del producto * precio del producto
            print(f"El valor del producto {diccionario_elemento['nombre']} es {valor_producto}") #printeamos el valor de cada producto
            valor_total += valor_producto #sumamos el valor de cada producto al valor total
        print(f"El valor total del inventario es {valor_total} euros")

    def realizar_compra(self):
        carrito = {} #creamos un diccionario vacío para el carrito de la compra
        total_compra = 0 # creamos una variable para ir sumando el valor de la compra
        while True: #Bucle while para que el cliente realice las compras
            self.ver_inventario() #mostrar el inventario para saber que productos están disponibles
            producto_tienda = input("Añada el producto a su carrito o escriba exit para salir de la compra") #preguntamos que producto quiere comprar 
            if producto_tienda.lower() == "exit":
                break #si introduce la palabra exit se termina la compra
            

            for diccionario_elemento in self.inventario:
                if diccionario_elemento["nombre"] == producto_tienda: #si el producto existe en nuestro inventario
                    cantidad_producto = int(input(f"¿Cuantas unidades de {producto_tienda} quieres comprar?"))
                    print(f"Has escogido {cantidad_producto} unidades del producto: {producto_tienda}")
                    if cantidad_producto > diccionario_elemento["cantidad"]: #aquí queremos saber si hay suficiente cantidad del producto seleccionado en nuestro invenetario
                        print(f"No tenemos stock suficiente de {producto_tienda} actualmente")
                        break #si no existe cantidad suficiente del producto
                    diccionario_elemento["cantidad"] = diccionario_elemento["cantidad"] - cantidad_producto #restamos la cantidad del producto que se ha introducido al carrito
                    total_ahora = cantidad_producto * diccionario_elemento["precio"] #variable del valor (en dinero) del producto por las unidades compradas
                    total_compra += total_ahora #actualizamos el precio del total de la compra
                    print(f"El valor de su carrito actualmente es de -> {total_compra} euros")

                    if producto_tienda in carrito: #si el cliente ya había seleccionado ese producto y está en el carrito
                        print(f"El producto {producto_tienda} ya se encuentra en el carrito")
                        carrito[producto_tienda]["cantidad"] += cantidad_producto #sumamos la cantidad nueva a la cantidad anterior en el carrito
                        carrito[producto_tienda]["total_compra"] += total_ahora #sumamos el total de este producto al total de la compra
                    elif diccionario_elemento["nombre"] != producto_tienda: #si el producto no se encuetra en nuestra tienda
                        print(f"El producto {producto_tienda} actualmente no se encuentra en nuestro stock")
                        
                    else:
                        carrito.update({producto_tienda:{"cantidad": cantidad_producto, "total_compra": total_ahora}}) #si el producto no se encuentra en el carrito lo añade
                
            
            print(f"Tu carrito se encuentra en el siguiente estado actualemente -> {carrito}")
            print(f"El valor total de tu compra es: {total_compra} euros")
            #printeamos carrito y coste total de la compra actualmente



    def agregar_cliente(self, nombre, email):
        if nombre in self.clientes: #vemos si ya existe el nombre
            print(f"El cliente {nombre} ya existe en la base de datos de la tienda")
        
        else:
            self.clientes[nombre] = {"nombre":nombre,"email":email} #si el nombre no existe lo añadimos al diccionario self.clientes
            print(f"El cliente {nombre} ha sido añadido a nuestra base de datos con el email: {email}")
            print(self.clientes) #printeamos el diccionario para comprobar que se ha añadido correctamente
        return
         
 
    def ver_clientes(self):
        for nombre, informacion in self.clientes.items(): # iteramos en el diccionario self.clientes y con items obtenemos la clave y el valor
            nombre = informacion["nombre"] #extraemos el nombre
            email = informacion["email"]#extraemos el email
            print(f"nombre: {nombre}, email: {email}")
            

                    
                    
                    
                 
                        
```

```python
tienda = TiendaOnline()
tienda.agregar_producto("pantalon" , 30 , 5)
tienda.agregar_producto("pantalon", 30, 20)
tienda.agregar_producto("camisa" , 15 , 5)
tienda.agregar_producto("calcetines", 3 , 50)
tienda.agregar_producto("falda" , 10 , 10)
tienda.agregar_producto("camiseta", 8, 32)
print("---------------------------------------------------------")
tienda.ver_inventario()
print("---------------------------------------------------------")
tienda.buscar_producto("pantalon")
tienda.buscar_producto("zapatos")
print("---------------------------------------------------------")
tienda.actualizar_stock("camisa" , 5)
print("----------------------------------------------------------")
tienda.eliminar_producto("falda")
print("-----------------------------------------------------------")
tienda.calcular_valor_inventario()
print("------------------------------------------------------------")
tienda.realizar_compra()




```
